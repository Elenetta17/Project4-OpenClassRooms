# 0.Imports and functions

```python
from category_encoders import TargetEncoder
from sklearn.compose import ColumnTransformer
from sklearn.dummy import DummyRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score,  mean_absolute_error, mean_squared_error
from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import PowerTransformer, QuantileTransformer
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import RobustScaler
from sklearn.svm import SVR
from xgboost import XGBRegressor

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import shap
import time

```

```python
def score(y_true, pred):
    """Cette fonction prend en entrée les données du jeu de test et les données prédites, et retourne un Dataframe avec 3 scores:
    R2, Mean Absolute Error (MAE) et Root Mean Squared Error (RMSE)"""
    #scores
    a = r2_score(y_true, pred)
    b = mean_absolute_error(y_true, pred)
    c = mean_squared_error(y_true, pred, squared=False)
    return pd.DataFrame([a,b,c], index = ['R2', 'MAE', 'RMSE'], columns = ['Scores'])


def inverse_variables(X_train, X_test):
    """Cette fonction prend en entrée un jeu de données d'entrainement et un jeu de données de test et y rajoute les inverses des variables"""
    inversed_cols = ['Inverse_' + col for col in X_train.columns] #nom des inverses des variables
    inversed_train = 1/(X_train+1) #calculs inverses sur le train set (le + 1 est pour ne pas avoir des valeurs infinis)
    inversed_train.columns = inversed_cols 
    inversed_test = 1/(X_test+1) #calculs inverses sur le test set (le + 1 est pour ne pas avoir des valeurs infinis)
    inversed_test.columns = inversed_cols
    X_train = pd.concat([X_train, inversed_train], axis = 1) #concatenation inverses train set
    X_test = pd.concat([X_test, inversed_test], axis = 1) #concatenation inverses test set
    return X_train, X_test

def cv_scores(estimator, X, y):
    """Cette fonction prend en entrée un estimator, un jeu de données et une variable cible, effectue une cross validation, imprime le runtime moyen et retourne un DataFrame
    avec R2, MAE et RMSE moyens sur les training folds et sur le validation fold"""
    results = cross_validate(estimator, X, y, cv=5, verbose = 3, scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'r2'], return_train_score=True)
    train_rmse = -results['train_neg_root_mean_squared_error'].mean()
    test_rmse = -results['test_neg_root_mean_squared_error'].mean()
    train_mae = -results['train_neg_mean_absolute_error'].mean()
    test_mae = -results['test_neg_mean_absolute_error'].mean()
    train_r2 = results['train_r2'].mean()
    test_r2 = results['test_r2'].mean()
    print('\nMean runtime: {:.3f}'.format(results['fit_time'].mean()))
    return pd.DataFrame([[train_r2, test_r2], [train_mae, test_mae], [train_rmse, test_rmse]], index = ['R2', 'MAE', 'RMSE'], columns=['Mean Train Folds', 'Mean Validation Fold'])
```

# 1.Encodage des données

Les variables catégorielles sont 3 : `PrimaryPropertyType`, `ZipCode` et `Neighborhood`.
Il faut les encoder pour pouvoir effectuer de modélisations. Le choix s'est porté sur le Target Encoding, car avec le OneHotEncoder j'aurais créé 84 variables et le risque d'overfitting aurait été agrandi.

```python
#lecture données
train = pd.read_csv('Data/energy_train.csv')
test = pd.read_csv('Data/energy_test.csv')
```

```python
#sélection variables catégorielles
categorical_columns = ['PrimaryPropertyType', 'ZipCode', 'Neighborhood']
categorical_columns
```

```python
#séparation en variables indépendantes et variable cible
y_train = train['SiteEUIWN(kBtu/sf)']
X_train = train.drop(['SiteEUIWN(kBtu/sf)'], axis = 1)
y_test = test['SiteEUIWN(kBtu/sf)']
X_test = test.drop(['SiteEUIWN(kBtu/sf)'], axis = 1)
```

## 3.1 Premières modélisations
