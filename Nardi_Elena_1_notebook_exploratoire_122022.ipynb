# 0.Imports et fonctions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
def homogenise_neighborhood(neighborhood):
    """Cette fonction permet d'homogénéiser la variable neighborhood"""
    if neighborhood == 'DELRIDGE NEIGHBORHOODS':
        return 'DELRIDGE'
    return neighborhood.upper()

def feature_engineering(train, test, name):
    """Cette fonction prend en entrée un train et test set, effectue le feature engineering sur les deux, et prepare et 
    sauvegarde les dataframes pour la modélisation"""
    
    train['Age'] = 2016 - train['YearBuilt']
    test['Age'] = 2016 - test['YearBuilt']
    train.drop('YearBuilt', axis = 1, inplace=True)
    test.drop('YearBuilt', axis = 1, inplace=True)
    
    train['Electricity_proportion'] = train['Electricity(kBtu)']/(train['Electricity(kBtu)'] + train['SteamUse(kBtu)'] + train['NaturalGas(kBtu)'])
    train['Steam_proportion'] = train['SteamUse(kBtu)']/(train['Electricity(kBtu)'] + train['SteamUse(kBtu)'] + train['NaturalGas(kBtu)'])
    train['NaturalGas_proportion'] = train['NaturalGas(kBtu)']/(train['Electricity(kBtu)'] + train['SteamUse(kBtu)'] + train['NaturalGas(kBtu)'])

    test['Electricity_proportion'] = test['Electricity(kBtu)']/(test['Electricity(kBtu)'] + test['SteamUse(kBtu)'] + test['NaturalGas(kBtu)'])
    test['Steam_proportion'] = test['SteamUse(kBtu)']/(test['Electricity(kBtu)'] + test['SteamUse(kBtu)'] + test['NaturalGas(kBtu)'])
    test['NaturalGas_proportion'] = test['NaturalGas(kBtu)']/(test['Electricity(kBtu)'] + test['SteamUse(kBtu)'] + test['NaturalGas(kBtu)'])
    
    train['Parking_proportion'] = train['PropertyGFAParking']/(train['PropertyGFABuilding(s)'] + train['PropertyGFAParking'])
    train['Building_proportion'] = train['PropertyGFABuilding(s)']/(train['PropertyGFABuilding(s)'] + train['PropertyGFAParking'])
    train['LargestUSEpct'] = train['LargestPropertyUseTypeGFA']/(train['PropertyGFABuilding(s)']+ train['PropertyGFAParking'])
    train['Side'] = np.sqrt(train['PropertyGFABuilding(s)']/((train['NumberofFloors'])*(train['NumberofBuildings'])))

    test['Parking_proportion'] = test['PropertyGFAParking']/(test['PropertyGFABuilding(s)'] + test['PropertyGFAParking'])
    test['Building_proportion'] = test['PropertyGFABuilding(s)']/(test['PropertyGFABuilding(s)'] + test['PropertyGFAParking'])
    test['LargestUSEpct'] = test['LargestPropertyUseTypeGFA']/(test['PropertyGFABuilding(s)']+ test['PropertyGFAParking'])
    test['Side'] = np.sqrt(test['PropertyGFABuilding(s)']/((test['NumberofFloors'])*(test['NumberofBuildings'])))
    
    train.drop(['SteamUse(kBtu)',
       'Electricity(kBtu)', 'NaturalGas(kBtu)', 'PropertyGFAParking', 'PropertyGFABuilding(s)',
       'LargestPropertyUseTypeGFA'],axis = 1, inplace=True)
    test.drop(['SteamUse(kBtu)',
       'Electricity(kBtu)', 'NaturalGas(kBtu)', 'PropertyGFAParking', 'PropertyGFABuilding(s)',
       'LargestPropertyUseTypeGFA'],axis = 1, inplace=True)
    
    data_energy_train = train.drop('GHGEmissionsIntensity',axis = 1)
    data_GHE_train = train.drop('SiteEUIWN(kBtu/sf)',axis = 1)
    data_energy_test = test.drop('GHGEmissionsIntensity',axis = 1)
    data_GHE_test = test.drop('SiteEUIWN(kBtu/sf)',axis = 1)
    
    data_energy_train.to_csv('Data/energy_train_' + name + '.csv', index = False)
    data_energy_test.to_csv('Data/energy_test_' + name + '.csv', index = False)
    data_GHE_train.to_csv('Data/GHE_train_' + name + '.csv', index = False)
    data_GHE_test.to_csv('Data/GHE_test_' + name + '.csv', index = False)
```

# 3. Feature engineering

Dans cette partie, je vais créer des variables potentiellement utiles pour la modélisation.

D'abord, je vais remplacer l'année de construction avec l'age de l'immeuble.

```python
#initialisation variable Age
train['Age'] = 2016 - train['YearBuilt']
test['Age'] = 2016 - test['YearBuilt']
```

```python
#suppression variable année de construction
train.drop('YearBuilt', axis = 1, inplace=True)
test.drop('YearBuilt', axis = 1, inplace=True)
```

Je ne vais pas garder les données sur la consommation d'électricité, gaz et vapeur telles qu'elles, mais je vais utiliser le pourcentage de chaque source d'énergie par rapport à la consommation totale. Ce pourcentage dépends des équipements nécessaires au fonctionnement de l'immeuble, et ne devrait pas beaucoup changer avec le temps.

```python
#initialization des variables "pourcentage de source"
train['Electricity_proportion'] = train['Electricity(kBtu)']/(train['Electricity(kBtu)'] + train['SteamUse(kBtu)'] + train['NaturalGas(kBtu)'])
train['Steam_proportion'] = train['SteamUse(kBtu)']/(train['Electricity(kBtu)'] + train['SteamUse(kBtu)'] + train['NaturalGas(kBtu)'])
train['NaturalGas_proportion'] = train['NaturalGas(kBtu)']/(train['Electricity(kBtu)'] + train['SteamUse(kBtu)'] + train['NaturalGas(kBtu)'])

test['Electricity_proportion'] = test['Electricity(kBtu)']/(test['Electricity(kBtu)'] + test['SteamUse(kBtu)'] + test['NaturalGas(kBtu)'])
test['Steam_proportion'] = test['SteamUse(kBtu)']/(test['Electricity(kBtu)'] + test['SteamUse(kBtu)'] + test['NaturalGas(kBtu)'])
test['NaturalGas_proportion'] = test['NaturalGas(kBtu)']/(test['Electricity(kBtu)'] + test['SteamUse(kBtu)'] + test['NaturalGas(kBtu)'])
```

```python
#statistiques de base sur les pourcentages de consommation
train[['Electricity_proportion', 'Steam_proportion', 'NaturalGas_proportion']].describe()
```

Etant donne que les variables cibles sont des consommations/émissions par unité de surface, à priori la surface ne devrait pas jouer un rôle fondamental dans la modélisation. Cependant, à partir des variables liées à la surface on peut extraire des informations sur les caractéristiques de l'immeuble.
Je vais créer les variables suivantes :
- pourcentage de surface occupée par le parking par rapport à la surface totale
- pourcentage de surface occupée par les bâtiments par rapport à la surface totale
- pourcentage de la surface destinée à l'usage principal par rapport à la surface totale
- "côté" de l'immeuble (la racine carrée de la surface de l'immeuble, si l'immeuble était carré)


```python
#initialisation variables
train['Parking_proportion'] = train['PropertyGFAParking']/(train['PropertyGFABuilding(s)'] + train['PropertyGFAParking'])
train['Building_proportion'] = train['PropertyGFABuilding(s)']/(train['PropertyGFABuilding(s)'] + train['PropertyGFAParking'])
train['LargestUSEpct'] = train['LargestPropertyUseTypeGFA']/(train['PropertyGFABuilding(s)']+ train['PropertyGFAParking'])
train['Side'] = np.sqrt(train['PropertyGFABuilding(s)']/((train['NumberofFloors'])*(train['NumberofBuildings'])))

test['Parking_proportion'] = test['PropertyGFAParking']/(test['PropertyGFABuilding(s)'] + test['PropertyGFAParking'])
test['Building_proportion'] = test['PropertyGFABuilding(s)']/(test['PropertyGFABuilding(s)'] + test['PropertyGFAParking'])
test['LargestUSEpct'] = test['LargestPropertyUseTypeGFA']/(test['PropertyGFABuilding(s)']+ test['PropertyGFAParking'])
test['Side'] = np.sqrt(test['PropertyGFABuilding(s)']/((test['NumberofFloors'])*(test['NumberofBuildings'])))
```

```python
#suppression variables remplacées
train.drop(['SteamUse(kBtu)', 'Electricity(kBtu)', 'NaturalGas(kBtu)', 'PropertyGFAParking', 'PropertyGFABuilding(s)',
       'LargestPropertyUseTypeGFA'], axis = 1, inplace=True)
test.drop(['SteamUse(kBtu)', 'Electricity(kBtu)', 'NaturalGas(kBtu)', 'PropertyGFAParking', 'PropertyGFABuilding(s)',
       'LargestPropertyUseTypeGFA'], axis = 1, inplace=True)
```
